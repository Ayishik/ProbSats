```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
y_mov = [1,-1]
N = 10
T = 30
paths = []
for n in range(N):
    curr = [0,20]
    path = [np.array(curr)]
    for t in range(1,T+1):
        y=curr[1]+random.choice(y_mov)
        curr = [t,y]
        path.append(np.array(curr))
        if y<=15 or y >=35:
            break
    paths.append(np.array(path))
paths = np.array(paths)
```

```python
plt.figure(figsize=(8,8))
plt.axis([0,30,12.5,37.5])
plt.plot([i for i in range((T+1)))],[15]*(T+1),'r',[i for i in range(T+1)],[35]*31,'r',[i for i in range(31)],[20]*31,'b')
for path in paths:
    plt.plot(path[:,0],path[:,1])
```
